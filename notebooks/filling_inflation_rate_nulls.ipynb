```python
import pandas as pd
```

```python
df_wide = pd.read_json('inflation_rate_wide_clean.json')
df_long = pd.read_json('inflation_rate_clean.json')
```

```python
df_wide.head()
```

```python
df_long.head()
```

```python
df_long.info()
```

```python
df_long_new = df_wide.melt(
    id_vars=["country_name", "country_code"],   # leaving unchanged
    var_name="year",                            # new column name
    value_name="inflation_rate"                 # the value we fill
)
```

```python
df_long_new.head()
```

```python
df_long_new.info()
```

```python
df_long_new["year"] = df_long_new["year"].astype(int)
```

```python
df_long[
    (df_long["country_name"] == "Lebanon") &
    (df_long["year"] == 2015)
] # should be missing
```

```python
df_long_new[
    (df_long_new["country_name"] == "Lebanon") &
    (df_long_new["year"] == 2015)
] # added successfully
```

```python
df_long_new["country_year"] = (
    df_long_new["country_code"] + df_long_new["year"].astype(str)
)
```

```python
df_long_new.head()
```

```python
repeat_order = ["country_name",
    "year",
    "inflation_rate",
    "country_code",
    "country_year"]
```

```python
df_reordered = df_long_new[repeat_order]
```

```python
df_reordered.head()
```

```python
df_reordered.to_json(
    "inflation_rate_clean.json",
    orient="records",
    indent=2,
    force_ascii=False
)
```
